
Now we have finished downloading all tweet data for each tweet_id in df using tweepy and stored in tweet_json.txt . Now we have to read that data and extract tweetID, retweet count and favourite count for all of the tweets and store in a dataframe.

```python
tweet_dict = []
with open('tweet_json.txt', mode='r') as f:
    for tweet_data in f:
        json_obj = json.loads(tweet_data)
        tweetID = json_obj['id']
        retweet_count = json_obj['retweet_count']
        fav_count = json_obj['favorite_count']
        
        tweet_dict.append({'tweet_id' : tweetID,
                           'retweet_count' : retweet_count,
                           'fav_count' : fav_count})
```

```python
df2 = pd.DataFrame(tweet_dict)
```
